```python
import tkinter as tk
from tkinter import messagebox
import psycopg2
from urllib.parse import urlparse

def get_connection():
    url = "postgresql://postgres:vz9cLFrdFtcdsb8i@db.kgnovzpgzrbhxcpfqjqq.supabase.co:5432/postgres"
    result = urlparse(url)
    return psycopg2.connect(
        database=result.path[1:],  
        user=result.username,
        password=result.password,
        host=result.hostname,
        port=result.port
    )

def add_student():
    first_name = entry_first_name.get()
    last_name = entry_last_name.get()
    date_of_birth = entry_date_of_birth.get()
    gender = entry_gender.get()
    enrollment_date = entry_enrollment_date.get()

    try:
        conn = get_connection()
        cursor = conn.cursor()
        cursor.execute(
            "INSERT INTO public.students (first_name, last_name, date_of_birth, gender, enrollment_date) VALUES (%s, %s, %s, %s, %s)",
            (first_name, last_name, date_of_birth, gender, enrollment_date)
        )
        conn.commit()
        messagebox.showinfo("Success", "Student added successfully")
        clear_entries()
    except Exception as e:
        messagebox.showerror("Error", str(e))
    finally:
        cursor.close()
        conn.close()

def clear_entries():
    entry_first_name.delete(0, tk.END)
    entry_last_name.delete(0, tk.END)
    entry_date_of_birth.delete(0, tk.END)
    entry_gender.delete(0, tk.END)
    entry_enrollment_date.delete(0, tk.END)

root = tk.Tk()
root.title("School Database System")
root.geometry("500x500")

frame = tk.Frame(root, padx=10, pady=10, bg="#f0f0f0")
frame.pack(padx=10, pady=10)

# tittel
title_label = tk.Label(frame, text="Add Student", bg="#f0f0f0", font=("Arial", 16))
title_label.grid(row=0, columnspan=2, pady=10)

# Ø¥input student information
tk.Label(frame, text="First Name", bg="#f0f0f0").grid(row=1, column=0, sticky="e")
entry_first_name = tk.Entry(frame, width=40)
entry_first_name.grid(row=1, column=1, pady=5)

tk.Label(frame, text="Last Name", bg="#f0f0f0").grid(row=2, column=0, sticky="e")
entry_last_name = tk.Entry(frame, width=40)
entry_last_name.grid(row=2, column=1, pady=5)

tk.Label(frame, text="Date of Birth (YYYY-MM-DD)", bg="#f0f0f0").grid(row=3, column=0, sticky="e")
entry_date_of_birth = tk.Entry(frame, width=40)
entry_date_of_birth.grid(row=3, column=1, pady=5)

tk.Label(frame, text="Gender", bg="#f0f0f0").grid(row=4, column=0, sticky="e")
entry_gender = tk.Entry(frame, width=40)
entry_gender.grid(row=4, column=1, pady=5)

tk.Label(frame, text="Enrollment Date (YYYY-MM-DD)", bg="#f0f0f0").grid(row=5, column=0, sticky="e")
entry_enrollment_date = tk.Entry(frame, width=40)
entry_enrollment_date.grid(row=5, column=1, pady=5)

#botton add student
add_button = tk.Button(frame, text="Add Student", command=add_student, bg="#4CAF50", fg="white", width=30)
add_button.grid(row=6, columnspan=2, pady=20)

root.configure(bg="#f0f0f0")
root.mainloop()
```
